# 1. What is Double Machine Learning?


Here's a simplified explanation of the Double Machine Learning method for causal inference:

### Modeling Nuisance Functions

We want to find out how a treatment $T$ affects an outcome $Y$ while controlling for other variables $X$ (confounders).

First, use machine learning to estimate two functions:
- $\hat{m}(X)$, which predicts $Y$ using $X$
- $\hat{g}(X)$, which predicts $T$ using $X$

### Orthogonalization (Removing Confounders)

Calculate the residuals, which are the parts of $Y$ and $T$ not explained by $X$:

$$
\tilde{Y} = Y - \hat{m}(X)
$$

$$
\tilde{T} = T - \hat{g}(X)
$$

These residuals represent $Y$ and $T$ with the effects of $X$ removed, making it easier to isolate the direct effect of $T$ on $Y$.

### Causal Effect Estimation

Run a simple linear regression of $\tilde{Y}$ on $\tilde{T}$:

$$
\theta = \text{argmin}_{\theta} \sum (\tilde{Y} - \theta \tilde{T})^2
$$

The coefficient $\theta$ from this regression gives us an unbiased estimate of the causal effect of $T$ on $Y$.

# 2. Demonstration

````python
import numpy as np
n = 1000
X = np.random.normal(0, 1, (n, 5)) 
T = np.random.binomial(1, 0.5, n)  
Y = 3 * T + np.dot(X, [1, -1, 0.5, 0, 0.2]) + np.random.normal(0, 0.5, n)  # Outcome
````

First, Setting Up Parameters
```python
n = 1000  # Number of samples to generate
```
- `n` represents the number of data points (observations) to create. In this example, we generate `1000` samples.

### 2. Generating Covariates \( X \)
```python
X = np.random.normal(0, 1, (n, 5))  # 1000 samples, 5 covariates/features
```
- **Explanation**:
  - `np.random.normal(0, 1, (n, 5))` generates a 2-dimensional array with `1000` rows and `5` columns.
  - Each entry in the array is sampled from a normal (Gaussian) distribution with a mean of `0` and a standard deviation of `1`.
- **Interpretation**:
  - This means that each of the 1000 samples has 5 features (covariates), which could represent different characteristics or attributes of the observations.

### 3. Generating Treatment Assignment \( T \)
```python
T = np.random.binomial(1, 0.5, n)  # Binary treatment assignment (0 or 1)
```
- **Explanation**:
  - `np.random.binomial(1, 0.5, n)` generates an array of `1000` binary values (0 or 1).
  - The parameters `1` and `0.5` specify that for each sample, a trial is performed where there is a 50% chance of getting `1` (treatment applied) and a 50% chance of getting `0` (no treatment).
- **Interpretation**:
  - The treatment assignment is essentially a binary indicator of whether an individual received the treatment or not. In real-world terms, this could represent, for example, whether a patient took a medication (1) or did not (0).

### 4. Generating the Outcome \( Y \)
```python
Y = 3 * T + np.dot(X, [1, -1, 0.5, 0, 0.2]) + np.random.normal(0, 0.5, n)  # Outcome
```
- **Explanation**:
  - The outcome \( Y \) is calculated based on:
    - A direct effect from the treatment: `3 * T`
      - This implies that when the treatment \( T \) is `1`, it increases \( Y \) by `3` units. If \( T \) is `0`, it has no effect.
    - A linear combination of the covariates \( X \): `np.dot(X, [1, -1, 0.5, 0, 0.2])`
      - `np.dot` computes the dot product between the matrix `X` (1000 samples × 5 features) and the weight vector `[1, -1, 0.5, 0, 0.2]`.
      - This assigns specific weights to the contribution of each covariate to the outcome:
        - The first covariate contributes with a weight of `1`.
        - The second covariate contributes negatively with a weight of `-1`.
        - The third covariate has a positive contribution of `0.5`.
        - The fourth covariate contributes `0` (i.e., it has no effect).
        - The fifth covariate contributes `0.2`.
    - Random noise: `np.random.normal(0, 0.5, n)`
      - Adds Gaussian noise to the outcome, sampled from a normal distribution with a mean of `0` and a standard deviation = 0.5/
      - This represents random variations or unobserved factors affecting the outcome.

- **Interpretation**:
  - The outcome \( Y \) depends on the treatment effect, a weighted combination of covariates, and some random noise. This reflects a real-world scenario where the outcome is influenced by both observed features and unobserved randomness.

### Complete Explanation of the Outcome Equation
The outcome \( Y \) for each observation is defined by:

Y = 3*T + (1*X1 - 1*X2 + 0.5 * X3 + 0*X4 + 0.2 * X5) + ϵ

Where:
- T is the binary treatment (either `1` or `0`).
- X1, X2, X3, X4, and X5 are the covariates/features.
- ϵ is the noise term drawn from a normal distribution with mean = 0 and SD = 0.5

### Intuition Behind the Data Generation
- **Covariates \( X \)**: Represent features that might influence both the treatment assignment and the outcome.
- **Treatment \( T \)**: Simulates a decision or intervention (e.g., receiving a drug, participating in a program).
- **Outcome \( Y \)**: The result that depends on the treatment and covariates, with added randomness to simulate real-life variability.

This synthetic data allows for testing causal inference models, such as double machine learning, to estimate the treatment effect \( 3 \) while controlling for the influence of covariates \( X \).

Data splitting is a crucial step in training machine learning models as it ensures that the model is evaluated on data that it has not seen before, helping to prevent overfitting and assess generalizability. In the context of double machine learning (DML), data splitting is used for training and testing purposes and may also be used in cross-fitting. Here is a detailed explanation of the data splitting code:


````python
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
X_train, X_test, T_train, T_test, Y_train, Y_test = train_test_split(X, T, Y, test_size=0.3, random_state=42)
````

### Detailed Explanation:
1. **`train_test_split` Function**:
   - This function from `scikit-learn` splits arrays or matrices into random train and test subsets.
   - It ensures that the training set is used to fit the model, and the test set is used to evaluate its performance.

2. **Parameters Explained**:
   - **`X`**: The covariates/features matrix (shape: \(n \times d\), where \(n\) is the number of samples and \(d\) is the number of features).
   - **`T`**: The treatment vector (shape: \(n\)), indicating whether the treatment was applied (binary or continuous).
   - **`Y`**: The outcome vector (shape: \(n\)), representing the response variable influenced by the treatment and covariates.

   - **`test_size=0.3`**:
     - Specifies that 30% of the data will be used as the test set and 70% as the training set.
     - This ensures a reasonable division where the model has enough data to learn from while retaining sufficient data for evaluation.

   - **`random_state=42`**:
     - This is a seed for the random number generator. By setting `random_state`, the function produces the same split every time the code is run, making the results reproducible.

3. **Output Variables**:
   - **`X_train, X_test`**:
     - `X_train`: The subset of the covariates used for training (70% of the original `X`).
     - `X_test`: The subset of the covariates used for testing (30% of the original `X`).

   - **`T_train, T_test`**:
     - `T_train`: The treatment vector used for training (70% of `T`).
     - `T_test`: The treatment vector used for testing (30% of `T`).

   - **`Y_train, Y_test`**:
     - `Y_train`: The outcome vector used for training (70% of `Y`).
     - `Y_test`: The outcome vector used for testing (30% of `Y`).

### Purpose of Data Splitting

#### Training the Model
The `X_train`, `T_train`, and `Y_train` sets are used to fit the DML model and train the nuisance functions ($\hat{Y}$ and $\hat{T}$).

#### Evaluation
The `X_test`, `T_test`, and `Y_test` sets are held out and used to evaluate the model's predictive performance and estimate the treatment effect on unseen data.

### Cross-Fitting Context:
In double machine learning, data splitting can also be used as part of cross-fitting:
- The training data can be further divided into subsets or "folds," where models are trained on one subset and tested on another. This helps minimize overfitting and biases in the estimated treatment effect.

### Visual Representation:
- **Training Set** (70% of data):
  - Used to train models, estimate residuals, and create nuisance functions.
- **Test Set** (30% of data):
  - Used for evaluating the final model performance, predicting treatment effects, and ensuring generalizability.

This approach ensures the reliability of the DML process by using separate data to train and test the model, ultimately leading to a more accurate and unbiased estimation of treatment effects.

````python
from sklearn.linear_model import LinearRegression
from econml.dml import LinearDML

estimator = LinearDML(
    model_y=LinearRegression(),  # Outcome model
    model_t=LinearRegression(),  # Treat binary treatment as regression
    random_state=42
)

# Fit the model using the training data
estimator.fit(Y_train, T_train, X=X_train)

print("Model fitting complete.")
````

This code sets up and fits a **double machine learning (DML)** model using `LinearDML` from the `econml` library. Below is a detailed explanation of each part:

### 1. **Import Statements**:
```python
from sklearn.linear_model import LinearRegression
from econml.dml import LinearDML
```
- **`from sklearn.linear_model import LinearRegression`**:
  - Imports the `LinearRegression` class from `scikit-learn`, a simple linear model used for regression analysis.
- **`from econml.dml import LinearDML`**:
  - Imports the `LinearDML` class from the `econml` library. `LinearDML` is a specific implementation of the double machine learning algorithm designed to estimate treatment effects while controlling for confounders.

### 2. **Model Initialization**:
```python
estimator = LinearDML(
    model_y=LinearRegression(),  # Outcome model
    model_t=LinearRegression(),  # Treatment model
    random_state=42
)
```
- **`estimator = LinearDML(...)`**:
  - Creates an instance of the `LinearDML` class. This object will be used to fit the model and estimate treatment effects.
- **`model_y=LinearRegression()`**:
  - Specifies that a `LinearRegression` model should be used to predict the outcome \( Y \) based on the covariates \( X \) and treatment \( T \). This is known as the **outcome model**.
- **`model_t=LinearRegression()`**:
  - Specifies that a `LinearRegression` model should be used to estimate the treatment \( T \) given the covariates \( X \). This is known as the **treatment model**. Although the treatment is binary, using linear regression here simplifies the treatment modeling as a regression problem.
- **`random_state=42`**:
  - Sets the seed for the random number generator to ensure reproducibility. The results will be consistent each time you run the code.

### 3. **Fitting the Model**:
```python
estimator.fit(Y_train, T_train, X=X_train)
```
- **`estimator.fit(...)`**:
  - Trains the `LinearDML` model using the provided training data.
- **`Y_train`**:
  - The actual outcomes or dependent variable for each observation in the training set.
- **`T_train`**:
  - The treatment variable for each observation, which indicates whether the treatment was applied (e.g., `0` or `1`).
- **`X_train`**:
  - The covariates or features that could affect both the treatment assignment and the outcome.

### 4. **Purpose of This Code**:
- The `LinearDML` model is used to estimate the **causal treatment effect** of \( T \) on \( Y \) while adjusting for confounding factors in \( X \).
- By fitting the model with `X_train`, `T_train`, and `Y_train`, `LinearDML` learns how the treatment and outcome relate to the covariates. This enables it to make unbiased estimates of how much the treatment affects the outcome (i.e., the treatment effect).

### 5. **Output**:
```python
print("Model fitting complete.")
```
- This line prints a confirmation message to indicate that the model has been successfully fitted with the training data.

````python
# Ensure you have a fitted model and test data (X_test)
treatment_effects = estimator.effect(X_test)  # Estimate treatment effects on the test set

# Calculate and display the ATE
ate = treatment_effects.mean()
print(f"Average Treatment Effect (ATE): {ate:.3f}")
````

The code snippet calculates and prints the **Average Treatment Effect (ATE)** based on the treatment effect estimates generated by the model.

### Explanation:
- **ATE** (Average Treatment Effect) represents the average effect of the treatment across all observations in your dataset. It quantifies how much, on average, the outcome \( Y \) would change if the treatment \( T \) were applied to everyone in the population compared to if it were not applied.
- **`treatment_effects`** is an array containing the estimated treatment effect for each observation in your test set. These estimates are generated using the `estimator.effect(X_test)` method of the `LinearDML` model or similar functions.

### Code Breakdown:
```python
ate = treatment_effects.mean()  # Calculate the mean of the estimated treatment effects
print(f"Average Treatment Effect (ATE): {ate:.3f}")  # Print the ATE formatted to 3 decimal places
```

- **`treatment_effects.mean()`**:
  - This line calculates the mean of the treatment effect estimates, which gives the ATE.
- **`print(f"Average Treatment Effect (ATE): {ate:.3f}")`**:
  - This line prints the ATE, formatted to three decimal places for readability.

### Interpretation:
- The **ATE** tells you the average change in the outcome that you can expect if the treatment is applied across the entire dataset.
- For example, if the ATE is `2.5`, it means that, on average, the treatment increases the outcome by `2.5` units.

### Why ATE is Important:
- **Policy Decisions**: Helps in understanding the overall impact of an intervention, which can guide decisions in policy-making, healthcare, or business strategies.
- **Causal Inference**: Provides a summary measure of the causal effect, helping to evaluate the effectiveness of a treatment or intervention.

This calculation is critical for summarizing the results of your causal analysis and understanding the broader implications of applying the treatment across a population.

````python
# Calculate the confidence intervals for the treatment effect estimates
lower_bound, upper_bound = estimator.effect_interval(X_test, alpha=0.05)  # 95% confidence interval

# Display the first few confidence intervals and the corresponding treatment effects
for i in range(5):
    print(f"Sample {i}: Treatment effect = {treatment_effects[i]:.3f}, "
          f"CI = [{lower_bound[i]:.3f}, {upper_bound[i]:.3f}]")

# Calculate and display the average treatment effect (ATE)
ate = treatment_effects.mean()
print(f"Average Treatment Effect (ATE): {ate:.3f}")

# Calculate p-values for the treatment effects (using z-scores)
import scipy.stats as stats

# Calculate standard errors from the confidence intervals
standard_errors = (upper_bound - lower_bound) / (2 * 1.96)  # For 95% CI, 1.96 is used for a normal distribution
````

### Explanation of the Code:

This code calculates and displays:
1. **Confidence intervals** for the treatment effect estimates.
2. **Average Treatment Effect (ATE)**.
3. **P-values** for the treatment effects to assess their statistical significance.

### Step-by-Step Explanation:

1. **Calculating Confidence Intervals**:
   ```python
   lower_bound, upper_bound = estimator.effect_interval(X_test, alpha=0.05)  # 95% confidence interval
   ```
   - **`estimator.effect_interval()`**: This method computes the lower and upper bounds of the confidence interval for the treatment effect estimates at the specified confidence level (95% in this case, with `alpha=0.05`).
   - The method returns two arrays: `lower_bound` and `upper_bound`, which represent the 95% confidence interval for each estimated treatment effect.

2. **Displaying Treatment Effects with Confidence Intervals**:
   ```python
   for i in range(5):
       print(f"Sample {i}: Treatment effect = {treatment_effects[i]:.3f}, "
             f"CI = [{lower_bound[i]:.3f}, {upper_bound[i]:.3f}]")
   ```
   - This loop prints the treatment effect estimates and their corresponding 95% confidence intervals for the first five samples.
   - **Interpretation**: Each output line shows the estimated treatment effect and the range within which the true effect is likely to fall with 95% confidence.

3. **Calculating and Displaying the ATE**:
   ```python
   ate = treatment_effects.mean()
   print(f"Average Treatment Effect (ATE): {ate:.3f}")
   ```
   - This calculates the **Average Treatment Effect (ATE)** as the mean of all treatment effect estimates. It provides a summary of the overall impact of the treatment across the dataset.

4. **Calculating P-values**:
   ```python
   import scipy.stats as stats

   # Calculate standard errors from the confidence intervals
   standard_errors = (upper_bound - lower_bound) / (2 * 1.96)  # For 95% CI, 1.96 is the z-value for 95%
   z_scores = treatment_effects / standard_errors
   p_values = 2 * (1 - stats.norm.cdf(np.abs(z_scores)))  # Two-tailed p-values
   ```
   - **Standard Errors**: Derived from the width of the confidence intervals. The formula `(upper_bound - lower_bound) / (2 * 1.96)` gives the standard error assuming a normal distribution.
   - **Z-scores**: Calculated as the treatment effect divided by the standard error. This indicates how many standard deviations the estimate is away from zero.
   - **P-values**: Calculated using the z-scores to assess the statistical significance of the treatment effect estimates. A small p-value (typically < 0.05) indicates strong evidence against the null hypothesis (i.e., that the treatment effect is zero).

5. **Displaying P-values**:
   ```python
   for i in range(5):
       print(f"Sample {i}: p-value = {p_values[i]:.5f}")
   ```
   - This loop prints the p-values for the treatment effect estimates of the first five samples.
   - **Interpretation**: A small p-value suggests that the treatment effect is statistically significant for that observation.

### Overall Interpretation:
- **Confidence Intervals** help determine the range within which the true treatment effect is likely to fall for each observation.
- The **ATE** provides an overall measure of the treatment's average effect.
- **P-values** indicate whether the estimated treatment effect is significantly different from zero for each observation.

````python
# Calculate treatment effects and confidence intervals
treatment_effects = estimator.effect(X_test)
lower_bound, upper_bound = estimator.effect_interval(X_test, alpha=0.05)

# Calculate the Average Treatment Effect (ATE)
ate = treatment_effects.mean()
print(f"Average Treatment Effect (ATE): {ate:.3f}")

# Calculate standard errors
standard_errors = (upper_bound - lower_bound) / (2 * 1.96)
ate_se = np.mean(standard_errors) / np.sqrt(len(X_test))

# Calculate statistical significance
z_score_ate = ate / ate_se
p_value_ate = 2 * (1 - stats.norm.cdf(np.abs(z_score_ate)))

# Calculate the confidence interval for the ATE
ci_lower_ate = ate - 1.96 * ate_se  # Lower bound for 95% CI
ci_upper_ate = ate + 1.96 * ate_se  # Upper bound for 95% CI

# Print results
print(f"Standard error (SE): {ate_se:.3f}")
print(f"z-score: {z_score_ate:.3f}")
print(f"p-value: {p_value_ate:.5f}")
print(f"95% Confidence Interval for ATE: [{ci_lower_ate:.3f}, {ci_upper_ate:.3f}]")
````

### Explanation of the Code:

This code snippet calculates the **standard error**, **z-score**, and **p-value** for the **Average Treatment Effect (ATE)**, providing an overall assessment of its statistical significance.

### Step-by-Step Explanation:

1. **Calculate the Standard Error for the ATE**:
   ```python
   ate_se = np.mean(standard_errors) / np.sqrt(len(X_test))  # Standard error of the mean ATE
   ```
   - **`standard_errors`**: The standard errors calculated earlier for each treatment effect estimate.
   - **`np.mean(standard_errors)`**: The mean standard error across all samples.
   - **`/ np.sqrt(len(X_test))`**: Divides by the square root of the number of samples to calculate the standard error of the mean ATE, ensuring that it represents the variability of the ATE estimate.

2. **Calculate the Z-score for the ATE**:
   ```python
   z_score_ate = ate / ate_se
   ```
   - The **z-score** is calculated as the ATE divided by its standard error. It measures how many standard deviations the ATE is from zero. A higher absolute z-score indicates that the ATE is further from zero, suggesting higher statistical significance.

3. **Calculate the P-value for the ATE**:
   ```python
   p_value_ate = 2 * (1 - stats.norm.cdf(np.abs(z_score_ate)))
   ```
   - **`stats.norm.cdf()`**: The cumulative distribution function of a standard normal distribution. This function calculates the probability that a standard normal variable is less than or equal to a given value.
   - The **p-value** is calculated using the z-score to determine the two-tailed probability of observing a value as extreme or more extreme than the observed ATE under the null hypothesis (which assumes the true ATE is zero).
   - **`2 * (1 - ...)`**: Multiplies by 2 to get a two-tailed p-value, which accounts for the possibility of the effect being significantly different in either direction (positive or negative).

4. **Display the Results**:
   ```python
   print(f"ATE: {ate:.3f}")
   print(f"Standard Error of ATE: {ate_se:.3f}")
   print(f"Z-score of ATE: {z_score_ate:.3f}")
   print(f"P-value of ATE: {p_value_ate:.5f}")
   ```

   - **ATE**: The average treatment effect calculated earlier.
   - **Standard Error of ATE**: Indicates the variability in the ATE estimate.
   - **Z-score of ATE**: Shows how far the ATE is from zero in terms of standard deviations.
   - **P-value of ATE**: Indicates whether the ATE is statistically significant. A small p-value (e.g., < 0.05) suggests strong evidence against the null hypothesis, indicating that the ATE is likely different from zero.

### Interpretation:
- **ATE**: The average effect of the treatment across the sample.
- **Standard Error**: A smaller standard error indicates more precise estimation of the ATE.
- **Z-score**: Higher absolute z-scores (typically > 1.96 for a 95% confidence level) imply the ATE is significantly different from zero.
- **P-value**: A small p-value (typically < 0.05) indicates that the ATE is statistically significant, meaning there is strong evidence that the treatment has an effect on the outcome.
